## Задание 1. Метод прямоугольников, метод трапеций.

```python
# Точность вычислений (знаков после запятой)
acc = 6

def RectMethod(y, a, b, count):
    step = (b - a) / count
    values = []
    for i in range(count):
        values.append(a + i * step)
    values.append(b)
    f_values = []
    min_val = 0
    max_val = 0
    for i in range(count):
        cur_f_val = y(values[i])
        min_val = min(cur_f_val, min_val)
        max_val = max(cur_f_val, max_val)
        f_values.append(cur_f_val)
    # Добавление прямоугольников на график
    visualization = plot(y, xmin = a, xmax = b, ymin = min_val, ymax = max_val, color = "red")
    for i in range(count):
        rect = [
            [values[i], 0],
            [values[i], f_values[i]],
            [values[i + 1], f_values[i]],
            [values[i + 1], 0]
        ]
        rect.append(rect[0])
        visualization += line(rect, color = "blue")
    # Вычисление ответа
    ans = 0
    for i in range(count):
        ans = ans + step * f_values[i]
#     При желании можно вывести интеграл
#     expr = LatexExpr(r'\int_{' + str(a.n(digits = acc)) + r'}^{' + str(b.n(digits = acc)) + r'} ' + latex(y(x)) + r' \,dx \approx ' + str(ans.n(digits = acc)))
    return visualization

def TrapMethod(y, a, b, count):
    step = (b - a) / count
    values = []
    for i in range(count):
        values.append(a + i * step)
    values.append(b)
    f_values = []
    min_val = 0
    max_val = 0
    for i in range(count + 1):
        cur_f_val = y(values[i])
        min_val = min(cur_f_val, min_val)
        max_val = max(cur_f_val, max_val)
        f_values.append(cur_f_val)
    # Добавление трапеций на график
    visualization = plot(y, xmin = a, xmax = b, ymin = min_val, ymax = max_val, color = "red")
    for i in range(count):
        trapeze = [
            [values[i], 0],
            [values[i], f_values[i]],
            [values[i + 1], f_values[i + 1]],
            [values[i + 1], 0]
        ]
        trapeze.append(trapeze[0])
        visualization += line(trapeze, color = "blue")
    # Вычисление ответа
    ans = 0
    for i in range(count):
        ans = ans + step * (f_values[i] + f_values[i + 1]) / 2
#     При желании можно вывести интеграл
#     expr = LatexExpr(r'\int_{' + str(a.n(digits = acc)) + r'}^{' + str(b.n(digits = acc)) + r'} ' + latex(y(x)) + r' \,dx \approx ' + str(ans.n(digits = acc)))
    return visualization

integral_f(x) = (ln(x-2) + 1) / (x-2)
left_integral_border = 2.5
right_integral_border = 50
min_steps = 1
max_steps = 100

@interact
def _(step = (min_steps, max_steps, 1)):
    visRect = RectMethod(integral_f, left_integral_border, right_integral_border, step)
    visTrap = TrapMethod(integral_f, left_integral_border, right_integral_border, step)
    graphics_array([visRect, visTrap]).show(frame=True, axes=False)
```

## Задание 2. Полиномы и расширенный алгоритм Евклида.

```python
# Расширенный алгоритм Евклида
# Возвращает НОД и коэффициенты
def __gcd(a, b):
    if (a == 0):
        return (b, 0, 1)
    q = b // a
    r = b % a
    d, x1, y1 = (i for i in __gcd(r, a))
    x = y1 - q * x1
    y = x1
    return (d, x, y)

# Вариант 2
R.<x> = QQ["x"]

f = 5*x^4 - 15*x^3 + 15*x^2 - 5*x
g = -22*x^5 + 33*x^4 - 11*x^3 - 99*x + 99

show(LatexExpr(r'\gcd(f(x), g(x)) = ' + latex(gcd(f, g))))
d, a, b = (i for i in __gcd(f, g))
show(LatexExpr(r'\gcd^{\sim}(f(x), g(x)) = ' + latex(d)))
```

## Задание 3. Упростить уравнение поверхности второго порядка в простраснтве.

```python
# Точность вычислений (знаков после запятой)
acc = 10

# Границы для трёхмерного графика
plot_range = 20.0

u(x, y, z) = 6*x^2 + 12*x*y + 7*y^2 + 2*x*z + 3*z^2 + 5*x + 5*y + 5*z - 18
show(u(x, y, z) == 0)
```

```python
# Из уравнения поверхности получим коэффициенты
a11 = 6
a22 = 7
a33 = 3

a12 = 6
a13 = 1
a23 = 0

a0 = -18
a1 = 5 / 2 
a2 = 5 / 2 
a3 = 5 / 2 

matrA = matrix([
    [a11, a12, a13],
    [a12, a22, a23],
    [a13, a23, a33]
])

vecA = vector([a1, a2, a3])

show(LatexExpr("A = " + latex(matrA)))
show(LatexExpr("a = " + latex(vecA)))
```

```python
# Характеристический многочлен
var("z")
matrAm = matrA - z * identity_matrix(3)
detA = det(matrAm).simplify_full()
show(LatexExpr(r'\det(A - z * E) = ' + latex(detA)))
```

```python
# Собственные числа
solutions = solve(detA == 0, z)
eigenvalues = []
for i in range(len(solutions)):
    if (solutions[i].rhs().imag().n(digits = acc) < 10 ** -acc):
        eigenvalues.append(solutions[i].rhs().real().n(digits = acc))
        show(LatexExpr("z_{" + str(i) + "} = " + str(eigenvalues[-1])))
```

```python
# Сравнение с методом eigenvalues
show(matrA.eigenvalues())
```

```python
# Собственные вектора
eigenvectors = []
var("x1 x2 x3")
for i in range(len(eigenvalues)):
    matrAmzE = matrA - eigenvalues[i] * identity_matrix(3)
    system_of_equations = []
    for j in range(3):
        system_of_equations.append(matrAmzE[j][0] * x1 + matrAmzE[j][1] * x2 + matrAmzE[j][2] * x3 == 0)
    # Решаем систему уравнений методом Гаусса
    system_of_equations[1] = system_of_equations[1] - system_of_equations[0] * matrAmzE[1][0] / matrAmzE[0][0]
    system_of_equations[2] = system_of_equations[2] - system_of_equations[0] * matrAmzE[2][0] / matrAmzE[0][0]
    system_of_equations[2] = system_of_equations[2] - system_of_equations[1] * (system_of_equations[2].lhs() / system_of_equations[1].lhs())
    # Проверяем, что ранг матрицы меньше трёх
    if (system_of_equations[2].lhs() == 0 and system_of_equations[2].rhs() == 0):
        system_of_equations[2] = (x3 == 1)
    else:
        print("Ранг матрицы равен трём")
    eigenvec = vector([j.rhs().n(digits = acc) for j in solve(system_of_equations, x1, x2, x3)[0]])
    ans = matrAmzE * eigenvec
    show(LatexExpr(r'\lambda_{' + str(i) + '} = ' + str(eigenvalues[i].n(digits = acc)) + ', s_{' + str(i) + '} = ' + str(eigenvec.n(digits = acc))))
    eigenvectors.append(eigenvec.n(digits = acc))
```

```python
# Сравнение с методом eigenvectors_right()
show(matrA.eigenvectors_right())
```

```python
# Получим матрицу перехода
matrS = matrix(eigenvectors)
show(LatexExpr("S = " + latex(matrS.n(digits = acc))))
# Нормируем собсветнные векторы
for i in range(len(eigenvectors)):
    eigenvectors[i] = eigenvectors[i] / sqrt(eigenvectors[i].dot_product(eigenvectors[i]))
matrS = matrix(eigenvectors)
# После нормировки получим
show(LatexExpr("S^{*} = " + latex(matrS.n(digits = acc))))
```

```python
matrAdiag = matrS * matrA * matrS.T
show(LatexExpr("S^{*} * A * {S^{*}}^T = " + latex(matrAdiag.n(digits = acc))))
```

```python
# Преобразуем вектор
vecAnew = matrS.T * vecA
show(vecAnew)
```

```python
v(x, y, z) = 0
variables = [x, y, z]
for i in range(len(variables)):
    v += matrAdiag[i][i] * variables[i] ^ 2 + variables[i] * vecAnew[i]
show(v(x, y, z) == 0)
```

```python
# Сделаем замену переменных
coef_linear = []

for i in range(len(vecAnew)):
    coef_linear.append((vecAnew[i] / (2 * matrAdiag[i][i])) ** 2)
    if (matrAdiag[i][i] * vecAnew[i] < 0):
        coef_linear[-1] = -coef_linear[-1]

v(x, y, z) = a0
variables = [x, y, z]
for i in range(len(variables)):
    v += matrAdiag[i][i] * (variables[i] + coef_linear[i]) ^ 2 - matrAdiag[i][i] * abs(coef_linear[i])
# После замены получим
show(v(x, y, z) == 0)
```
